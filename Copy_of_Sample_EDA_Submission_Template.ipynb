## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# 1. check and then correct the data types of all the parameter
df.dtypes       # every parameter data type is correct
````

````python
# 2. drop rows with very less missing values
df.dropna(subset = ['country'],axis = 0 , inplace=True)
````

````python
# replace the mean value in the column which have very large data is missing
mean_value = df["company"].mean()
df["company"].fillna(value=mean_value,inplace=True)

mean_value1 = df["agent"].mean()
df["agent"].fillna(value=mean_value1,inplace=True)
````

````python
# now dataset has no missing value
sns.heatmap(df.isna())
````

````python
# 3. identify the duplicates
df.duplicated().value_counts()
````

````python
# drop the duplicates
df.drop_duplicates(inplace=True)
````

````python
#4. Remove Outliers
sns.boxplot(x="adr",data=df)
````

````python
Q1 = df["adr"].quantile(.25)
Q3 = df["adr"].quantile(.75)
IQR = Q3 - Q1
````

````python
lower = Q1 - 1.5*IQR
higher = Q3 + 1.5*IQR
````

````python
df = df[(df["adr"]>=lower) & (df["adr"]<=higher)]
````

````python
# now dataset has no outliers
sns.boxplot(x="adr",data=df)
````

### What all manipulations have you done and insights you found?

Data wrangling is the process of removing errors and combining complex data sets to make them more accessible and easier to analyze

1. Firstly, in the dataset we make sure all the parameter in the dataset are in their desired types.
2. We identify the null values , the column which have very less amount of missing data relative to full table, in that case we will delete rows for that column.
3. The column which have very high amount of missing data , in that case we will replace all the dat with the mean value of that column data.
4. Now , we will identify the duplicates rows in the table and then we will delete the duplicates row.And we will confirm that by heat map.
5. Now, we will remove the outliers of the main column by using the boxpot , we will identify outliers and then we will remove outliers.
6. Now our dataset is clean and ready for the analysis.
